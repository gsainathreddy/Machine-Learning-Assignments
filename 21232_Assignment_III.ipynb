Splitting data for training and testing


```python
from sklearn.model_selection import train_test_split
x = X.iloc[:,:-1]
y = X.iloc[:,4]
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=39)
X_train.shape



```

Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logistic=LogisticRegression()
logistic.fit(X_train,y_train)
prediction=logistic.predict(X_test)
prediction
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction,y_test))
print('The Precision of the Logistic Regression is',metrics.precision_score(prediction,y_test,average='micro'))
print('The recall of the Logistic Regression is',metrics.recall_score(prediction,y_test,average='micro'))
print('The f1  of the Logistic Regression is',metrics.f1_score(prediction,y_test,average='micro'))
```

```python
from mlxtend.plotting import plot_decision_regions
value=0
width=1

plot_decision_regions(X_train.values, y_train.values, logistic, feature_index=[2,3], filler_feature_values={0: value, 1:value},
                      filler_feature_ranges={0: width, 1: width}, legend=3)
plt.ylabel('Petal Width (standardized)')
plt.xlabel('Petal Length (standardized)')
plt.title("Decision Boundary produced by Logistic Regression Model", size = 14)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train,y_train)
prediction=gnb.predict(X_test)

print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction,y_test))
print('The precision of the Logistic Regression is',metrics.precision_score(prediction,y_test,average='micro'))
print('The recall of the Logistic Regression is',metrics.recall_score(prediction,y_test,average='weighted'))
print('The f1 of the Logistic Regression is',metrics.f1_score(prediction,y_test,average='weighted'))
```

```python
## Decision Boundary
plot_decision_regions(X_train.values, y_train.values, gnb, feature_index=[2,3], filler_feature_values={0: value, 1:value},
                      filler_feature_ranges={0: width, 1: width}, legend=3)
plt.ylabel('Petal Width (standardized)')
plt.xlabel('Petal Length (standardized)')
plt.title("Decision Boundary produced by Naive Bayes Model", size = 14)
```
